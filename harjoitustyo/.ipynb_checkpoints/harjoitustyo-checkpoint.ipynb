#### Tehtävä 7. Pura tämän kansion sisällä olevan `data.html` -tiedoston `table`-elementin solujen arvot käyttäen regex-syntaksia ja yhdistä ne vaiheessa 1 luotuun dataframeen. Huom: tehtävässä tulee käyttää vain regexiä (Esim. Pandasin read_html()-metodin käyttö on kielletty)!

```python
import re
```

```python
f = open("data.html", "r")
file=f.read()
print(file)
```

```python
column_names=[]
re_pattern=r'\bth.*\b'
re_search = re.findall(re_pattern, file)
re_search
column_names.append(re_search)

def clean_colnames(row_elems):
    clean_elems=[]
    for elem in row_elems:
#         print(elem)
        clean_elems.append(re.sub('th>|</th|th, td','',elem))
    clean_elems=[x for x in clean_elems if x != '']
    return clean_elems

def get_clean_colnames(text):
    re_pattern=r'\bth.*\b'
    colnames = re.findall(re_pattern, text)
    cleaned_colnames=[]
#     print(colnames)
    cleaned_colnames=clean_colnames(colnames)
#     for colname in colnames:
#         cleaned_colnames.append(clean_colnames(colname))
    return cleaned_colnames

get_clean_colnames(file)
```

```python
import pandas as pd
#td=[]
re_pattern_td="<tr[^>]*>[\s\S]*?</tr>\s*"
re_search1=re.findall(re_pattern_td, file)
re_search1

def get_rows(text, re_pattern_td="<tr[^>]*>[\s\S]*?</tr>\s*"):
    rows=re.findall(re_pattern_td, text)
    return rows

def get_row_elems(row, re_pattern='<td[^>]*>[\s\S]*?</td>\s*'):
    row_elems = re.findall(re_pattern, row)
    return row_elems

def clean_elem(row_elem):
    cleaned_elem=re.sub('<td>\n|</td>\n|\n|','',row_elem)
    cleaned_elem=re.sub(' +', '', cleaned_elem)
    return cleaned_elem

def clean_elems(row_elems):
    cleaned_row_elems=[]
    for elem in row_elems:
        cleaned_row_elems.append(clean_elem(elem))
    return cleaned_row_elems

def clean_rows(rows):
    cleaned_rows=[]
    for i, row in enumerate(rows):
        if i==0:
            cleaned_rows.append(get_clean_colnames(row))
        else:
            row_elems=get_row_elems(row)
            cleaned_rows.append(clean_elems(row_elems))
    return cleaned_rows

rows=get_rows(file)
elems1=get_row_elems(rows[1])

# clean_elems(elems1)
rows_list=clean_rows(rows)
df=pd.DataFrame(rows_list[1:], columns=rows_list[0])
df.head()
# df.rename(columns=df.iloc[0])
# for elem in elems1:
#     print(clean_elem(elem))
# rows[1]
# re.findall('<td[^>]*>[\s\S]*?</td>\s*', rows[1])
```
